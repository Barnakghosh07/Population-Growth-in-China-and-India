# Population Growth in China and India

```python
#Compare the population growth in the China and India

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('countries.csv')
china = data[data.country == 'China']
india = data[data.country == 'India']

plt.plot(china.year, china.population / china.population.iloc[0] *100)
plt.plot(india.year, india.population / india.population.iloc[0] *100)

plt.legend(["China", "India"])
plt.xlabel('Year',fontsize=15)
plt.ylabel('Population Growth',fontsize=15)
plt.show()
```
